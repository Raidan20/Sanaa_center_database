```python
import pandas as pd
import datetime

#Editing the dates of the database of the web
data = pd.read_csv(r"E:\Sanaa Center\04_ Other Tasks\02_ All weekly reports database\Web Page\data\Database_v02.csv")

formated_dates = []
dates = list(data['Date'].values)
for date in dates:
    if pd.isna(date):
        formated_dates.append(None)
    else:
        try:
            date_split = date.split(' ')
            print(date_split)
            if '00:00' in date_split:
                date_split.remove('00:00')
                print(date_split)
                print('-----------')
                date = date_split[0]
                print(date)
            date = datetime.datetime.strptime(date, '%m/%d/%y')
        except:
            date = datetime.datetime.strptime(date, '%m/%d/%Y')
        formated_dates.append(date)

indx = 0
for index, row in data.iterrows():
    data.loc[index, 'Date'] = formated_dates[indx]
    indx = indx + 1

data.to_csv(r"E:\Sanaa Center\04_ Other Tasks\02_ All weekly reports database\Web Page\data\Database_v02.csv", index=False)
```

```python
import pandas as pd
import json

# Step 1: Read the CSV file
csv_file = r"Data\Database_v04.csv"  # Replace with your file path
df = pd.read_csv(csv_file)

# Step 2: Convert DataFrame to a list of dictionaries (JSON-like structure)
data_list = df.to_dict(orient="records")

# Step 3: Create a JavaScript file with the data
js_variable_name = "eventsData"
js_file = r"Data\Database_v02.js"  # Output JavaScript file

# Step 4: Write to the .js file
with open(js_file, "w") as f:
    f.write(f"const {js_variable_name} = ")
    json.dump(data_list, f, indent=4)  # Convert to JSON and write
    f.write(";")  # End with a semicolon to complete the JavaScript syntax

print(f"JavaScript file '{js_file}' created successfully!")
```

```python
csv_file = r"Data\Database_v03.csv"  # Replace with your file path
df = pd.read_csv(csv_file)

E = []
N = []
for index, row in df.iterrows():
    if pd.isna(row[7]):
        E.append(None)
        N.append(None)
    else:
        cords = row[7].split(',')
        E.append(cords[0])
        N.append(cords[1])

df['E'] = E
df['N'] = N

df.to_csv(r"Data\Database_v03_QGIS.csv", index=False)              
```
